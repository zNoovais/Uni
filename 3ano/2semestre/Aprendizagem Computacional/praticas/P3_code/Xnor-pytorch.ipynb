```python
## Example of XNOR with PyTorch
```

```python
## !pip install torch
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
## Dataset: XNOR truth table
X = torch.tensor([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
])
y = torch.tensor([
    [1.],
    [0.],
    [0.],
    [1.]
])
```

```python
# Neural network model
class XNORNet(nn.Module):
    def __init__(self, hidden = 2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(2, hidden),   # hidden layer - 4 nodes
            nn.Sigmoid(),      # nonlinearity (important!)
            nn.Linear(hidden, 1),   # output layer - 1 node
            nn.Sigmoid()       # output as probability
        )

    def forward(self, x):
        return self.net(x)

model = XNORNet(4)
```

```python
print(model)
```

```python
# Alternative way of defining the same model:
class XNORNet_v2(nn.Module):
    def __init__(self, hidden=2):
        super().__init__()

        self.fc1 = nn.Linear(2, hidden)
        self.act1 = nn.Sigmoid()

        self.fc2 = nn.Linear(hidden, 1)
        self.act2 = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        return x
```

```python
# Loss and optimizer
criterion = nn.BCELoss() # binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.05) # training algorithm. 
# model.parameters() are the weights and biases of the model that will be updated during training
```

```python
# Training loop
epochs = 100
for epoch in range(epochs):
    # forward
    y_pred = model(X)   # get predictions from the model
    loss = criterion(y_pred, y)     # compute the loss between predictions and true labels

    # backward
    optimizer.zero_grad()   # reset gradients to zero before backpropagation, so that they don't accumulate across epochs
    loss.backward()  # compute gradients of the loss with respect to model parameters
    optimizer.step()    # update model parameters using the computed gradients

    # print progress
    if epoch % 20 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss.item():.4f}")
```

```python
# Test the trained model
with torch.no_grad(): # no need to record operations/store intermediate tensors for backpropagation during predictions
    preds = model(X)
    print("\nPredictions:")
    print(preds.round())
```
