## ElGamal 
$p$ primo, $r$ raiz primitiva de $p$, $1\le a \le p-1$, $b \equiv r^a \mod  p$

ChPub: $(p, r, b)$
ChPriv: $a$

Cifrar $mens$ $(\gamma, \delta)$
$$\gamma = r^k \mod p$$
$$delta = mens\cdot b^k \mod p$$

Decifrar $\left(\gamma^a\right)^{-1}\cdot \delta \mod p$
